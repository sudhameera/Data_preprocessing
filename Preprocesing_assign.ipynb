```python
import pandas as pd
import numpy as np
```

```python
data=pd.read_csv('kidney_disease.csv')
```

```python
data.info()
```

```python
data
```

```python
data.isna().sum()
```

```python
data.columns
```

```python
qual=[]
quan=[]

for columnName in data.columns:
    if(data[columnName].dtype=='O'):
        qual.append(columnName)
    else:
        quan.append(columnName)
    
```

```python
qual
```

```python
quan
```

```python
data[quan]
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy='mean')
imp.fit(data[quan])
```

```python
data_quan=imp.transform(data[quan])
```

```python
data_quan
```

```python
data_quan=pd.DataFrame(data_quan,columns=quan)
```

```python
data_quan
```

```python
data_quan.isnull().sum()
```

```python
from sklearn.impute import SimpleImputer
imp=SimpleImputer(missing_values=np.nan, strategy='most_frequent')
imp.fit(data[qual])
```

```python
data_qual=imp.transform(data[qual])
```

```python
data_qual
```

```python
data_qual=pd.DataFrame(data_qual,columns=qual)
data_qual
```

```python
data_qual.isna().sum()
```

```python
processed_data=pd.concat([data_qual,data_quan],axis=1)
```

```python
processed_data
```

```python
processed_data.isnull().sum()
```

```python
processed_data.info()
```

```python
processed_data=pd.concat([data_quan,data_qual],axis=1)
```

```python
processed_data.info()
```

```python
processed_data.to_csv("preprocessed_data.csv",index=False)
```
